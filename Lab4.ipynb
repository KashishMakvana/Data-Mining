### 1) Write a python program to compute distance between Given two objects represented by the tuples (22, 1, 42, 10) and (20, 0, 36, 8): 
(a) Compute the Euclidean distance between the two objects. <BR>
(b) Compute the Manhattan distance between the two objects. <BR>
(c) Compute the Minkowski distance between the two objects, using q = 3. <BR>
(d) Compute the supremum distance between the two objects.<BR>


```python
import math as m
x = (22,1,42,10)
y = (20,0,36,8)
sum = 0 
for i in range(len(x)):
    sum += (x[i]-y[i])**2

print(f"Eucllidean distance : {m.sqrt(sum)}")
```

```python
import math as m
x = (22,1,42,10)
y = (20,0,36,8)
sum = 0 
for i in range(len(x)):
    sum += (x[i]-y[i])
print(f" Manhattan distance : {abs(sum)}")
```

```python
import math as m
x = (22,1,42,10)
y = (20,0,36,8)
sum = 0 
q = 3
for i in range(len(x)):
    sum += (abs(x[i]-y[i])**q)
print(f"Minkowski distance : {sum**(1/q)}")
```

```python
import math as m
x = (22,1,42,10)
y = (20,0,36,8)
sum = 0
for i in range(len(x)):
    sum = max(sum,abs(x[i]-y[i]))
print(f"Supremum distance : {sum}")
```

### 3) Write programs to perform the following tasks of preprocessing.

Equal Width Binning <BR>
Equal Frequency/Depth Binning

```python
import numpy as np
```

```python
data = [5,10,11,13,15,35,50,55,72,92,204,215]
df = pd.DataFrame(data,columns=['Value'])
num_bins = 3
bin_edges = np.linspace(df['Value'].min(),df['Value'].max(),num_bins+1)
df['Equal_Width_Bin'] = pd.cut(df['Value'],bins=bin_edges,labels=False,include_lowest=True)
print('Equal Width Binning:\n',df)
```

```python
data = [5,10,11,13,15,35,50,55,72,92,204,215]
df = pd.DataFrame(data,columns=['Value'])
num_bins = 3
no_of_data = len(data)
points_in_bin = no_of_data/num_bins
for i in range(0,len(data),int(points_in_bin)):
    print('Equal Depth Binning:\n',data[i:i+int(points_in_bin)])
```
